```python
def standardize_number(number: str) -> str:
    """
    Стандартизирует номер телефона.
    Удаляет пробелы и дефисы, заменяет номер, начинающийся на 8, на формат +7.

    :param number: Номер телефона (строка).
    :return: Стандартизированный номер (строка).
    """
    number = number.replace(" ", "").replace("-", "")
    if number.startswith('8'):
        number = "+7" + number[1:]
    elif not number.startswith('+'):
        number = "+7" + number
    return number

def standardize_name(name: str) -> str:
    """
    Стандартизирует имя.
    Приводит первую букву каждого слова к верхнему регистру.

    :param name: Имя (строка).
    :return: Стандартизированное имя (строка).
    """
    return " ".join(word.capitalize() for word in name.split())

phone_book={}

def add_contact() -> None:
    """
    Добавляет новый контакт в телефонную книгу.

    Запрашивает у пользователя имя и номер телефона,
    затем добавляет их в телефонную книгу.
    """
    name = standardize_name(input("Введите название для контакта: "))
    number = standardize_number(input("Введите номер телефона: "))
    phone_book[name] = number
    print("Контакт", name, "успешно добавлен!")

def delete_contact() -> None:
    """
    Удаляет контакт из телефонной книги.

    Запрашивает у пользователя имя контакта для удаления
    и удаляет его из телефонной книги, если он существует.
    """
    name = standardize_name(input("Введите имя контакта для удаления: "))
    if name in phone_book:
        del phone_book[name]
        print("Контакт", name, "успешно удалён!")
    else:
        print("Контакт не найден!")

def show_contacts() -> None:
    """
    Показывает все контакты в телефонной книге.

    Если телефонная книга пуста, выводит соответствующее сообщение.
    """
    if phone_book:
        print("Список контактов:")
        for name, number in phone_book.items():
            print(name + ":", number)
    else:
        print("Телефонная книга пуста.")

def change_number() -> None:
    """
    Изменяет номер телефона для существующего контакта.

    Запрашивает у пользователя имя контакта и новый номер,
    затем обновляет номер в телефонной книге.
    """
    name = standardize_name(input("Введите имя контакта для изменения номера: "))
    if name in phone_book:
        new_number = standardize_number(input("Введите новый номер телефона: "))
        phone_book[name] = new_number
        print("Номер для", name, "успешно изменён!")
    else:
        print("Контакт не найден!")

def main_menu() -> None:
    """
    Главное меню телефонной книги.

    Позволяет пользователю выбирать действия: добавление, удаление,
    просмотр контактов, изменение номера или выход из программы.
    """
    while True:
        print("Выберите функцию:")
        print("1. Добавить контакт")
        print("2. Удалить контакт")
        print("3. Показать список контактов")
        print("4. Изменить номер")
        print("5. Выход")
        choice = input("Введите номер функции: ")
        if choice == "1":
            add_contact()
        elif choice == "2":
            delete_contact()
        elif choice == "3":
            show_contacts()
        elif choice == "4":
            change_number()
        elif choice == "5":
            print("Выход из программы.")
            break
        else:
            print("Неверный выбор, попробуйте снова.")
main_menu()
```
